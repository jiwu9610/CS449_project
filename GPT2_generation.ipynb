```python
import torch
from transformers import pipeline, GPT2LMHeadModel, GPT2Tokenizer, BertForPreTraining, BertTokenizer, BertLMHeadModel
from google.colab import drive

# Mount Google Drive
drive.mount('/content/gdrive')
```

```python
prompts = [
    "The burden of decision in a criminal case lies with",
    "In a criminal trial, the prosecution must prove the defendant's guilt",
    "Unless it was a written check, I'm guessing it was an electronic transaction which means it is probably covered by EFTA. If so, the burden is on the bank to", #business law
    "Lawncare company provided service without expressed consent or prior payment, so I am"  # business law
]


question_prompts = [
    "Is Amazon or other E-commerce companies required by US law to refund a product that was lost in transit regardless of how long the request was made?" # business law
]
```

```python
#GPT2 Generation
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.add_special_tokens({"pad_token": "<pad>", 
                                "bos_token": "<startofstring>",
                                "eos_token": "<endofstring>"})
tokenizer.add_tokens(["<bot>:"])


model = GPT2LMHeadModel.from_pretrained("gpt2")
model.resize_token_embeddings(len(tokenizer))
state_dict = torch.load('/content/gdrive/My Drive/saved_model/model_GPT2_full_new.pth')
state_dict = {k:v for k,v in state_dict.items() if k not in ["transformer.h.0.attn.bias", "transformer.h.0.attn.masked_bias", "transformer.h.1.attn.bias",
                                                             "transformer.h.1.attn.masked_bias", "transformer.h.2.attn.bias", "transformer.h.2.attn.masked_bias", 
                                                             "transformer.h.3.attn.bias", "transformer.h.3.attn.masked_bias", "transformer.h.4.attn.bias",
                                                             "transformer.h.4.attn.masked_bias", "transformer.h.5.attn.bias", "transformer.h.5.attn.masked_bias",
                                                             "transformer.h.6.attn.bias", "transformer.h.6.attn.masked_bias", "transformer.h.7.attn.bias",
                                                             "transformer.h.7.attn.masked_bias", "transformer.h.8.attn.bias", "transformer.h.8.attn.masked_bias",
                                                             "transformer.h.9.attn.bias", "transformer.h.9.attn.masked_bias", "transformer.h.10.attn.bias",
                                                             "transformer.h.10.attn.masked_bias", "transformer.h.11.attn.bias", "transformer.h.11.attn.masked_bias"]}
model.load_state_dict(state_dict)

text_generation = pipeline("text-generation", model=model, tokenizer = tokenizer, pad_token_id=50256)

# prefix_text = prompts[0]
# generated_text= text_generation(prefix_text, max_length=200, do_sample=False)[0]
# print(generated_text['generated_text'])



for prefix_text in prompts + question_prompts:
  print("\n----------------------------------------------------------------------------------------------------------------------------------------------------")
  print(f"Prompt: {prefix_text}")
  generated_text= text_generation(prefix_text, max_length=100, do_sample=False)[0]

  print("Response:")
  print(generated_text['generated_text'].split('\n')[0])
```
