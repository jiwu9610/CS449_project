```python
from transformers import BertForSequenceClassification, BertTokenizer, BertModel
from google.colab import drive

drive.mount('/content/gdrive')

drive_path = '/content/gdrive/My Drive/saved_model/'
file_path = 'model_legalbert_full_more_data.pth'

model_path = drive_path + file_path
tokenizer = BertTokenizer.from_pretrained('nlpaueb/legal-bert-small-uncased') 

model = BertModel.from_pretrained('nlpaueb/legal-bert-small-uncased')
model.load_state_dict(torch.load(model_path), strict = False)  

# Freeze all the layers
for param in model.parameters():
    param.requires_grad = False

classification_model = BertForSequenceClassification(model.config)

classification_model.bert = model

num_classes = 14 
classification_model.classifier = torch.nn.Linear(768, num_classes)

file_path = 'model_legalbert_LayerFreeze.pth'

folder_path = 'legalbert_LayerFreeze'

# Save the model
model.save_pretrained(drive_path+folder_path, from_pt=True)
```
